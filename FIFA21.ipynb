```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data_fifa21.csv')
df
```

```python
#Do the height and weight columns have the appropriate data types?


df = df.rename(columns = {'Height':'Height(cm)','Weight':'Weight(Kg)'})



def quitar_caracteres (data):
    return data[:-2]

df['Height(cm)']=df['Height(cm)'].apply(quitar_caracteres)
df['Weight(Kg)']=df['Weight(Kg)'].apply(quitar_caracteres)


df['Weight(Kg)'] = pd.to_numeric(df['Weight(Kg)'],errors='coerce')
df['Height(cm)'] = pd.to_numeric(df['Height(cm)'],errors='coerce')

df[['Height(cm)','Weight(Kg)']]

```

```python
#Can you separate the joined column into year, month and day columns?

#SEPARANDOLO COMO STRING
df['Year_Joined'] = df['Joined'].str.split(', ' ,expand=True)[1].str.split('.', expand=True)[0]
df['Month_Joined'] = df['Joined'].str.split(' ', expand=True)[0]
df['Day_Joined'] = df['Joined'].str.split(' ' ,expand=True)[1].str.split(',', expand=True)[0]

df[['Joined','Year_Joined','Month_Joined','Day_Joined']]



#SEPARANDOLO COMO FORMATO DATE
# Supongamos que tu DataFrame se llama df y tu columna se llama 'fecha'
df['fecha_joined'] = pd.to_datetime(df['Joined'], format='%b %d, %Y')
# Crear una nueva columna con solo el mes
df['mes'] = df['fecha_joined'].dt.month
# Crear una nueva columna con solo el día
df['dia'] = df['fecha_joined'].dt.day
# Crear una nueva columna con solo el año
df['anio'] = df['fecha_joined'].dt.year




```

```python
#Can you clean and transform the value, wage and release clause columns into columns of integers.

df[['Value','Wage','Release Clause']]


def transformar_s (data):
    if(data[-1]=='K'):
        data = data[:-1]
        data = data[1:]
        data = pd.to_numeric(data,errors='coerce')
        data = data * 1000
        return data
    if(data[-1]=='M'):
        data = data[:-1]
        data = data[1:]
        data = pd.to_numeric(data,errors='coerce')
        data = data * 1000000
        return data
    else:
        data = data[1:]
        data = pd.to_numeric(data,errors='coerce')




df['Wage'] = df['Wage'].apply(transformar_s)
df['Wage'] = df['Wage'].fillna(0)
df['Wage']= df['Wage'].astype(np.int64)

df['Value'] = df['Value'].apply(transformar_s)
df['Value'] = df['Value'].fillna(0)
df['Value']= df['Value'].astype(np.int64)

df['Release Clause'] = df['Release Clause'].apply(transformar_s)
df['Release Clause'] = df['Release Clause'].fillna(0)
df['Release Clause']= df['Release Clause'].astype(np.int64)

df[['Wage','Value','Release Clause']]

```

```python
#Remove \n characters from Club



def remove_characters(data):
    return data[4:]

df['Club'] = df['Club'].apply(remove_characters)
df[['Club','Contract']]
```

```python
#Should you separate the team and contract column into separatem team and contract columns?


df[['start_contract','end_contract']] = df['Contract'].str.split(' ~ ', expand = True)
df.drop(columns=['Contract'])
df[['start_contract','end_contract','Club']]
```
